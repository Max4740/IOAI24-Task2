```python
!pip install ultralytics
!pip install -U ipywidgets
from ultralytics import YOLO
import pandas as pd
import numpy as np
```

```python
model = YOLO('yolov8l-cls.pt')
```

```python
result_model=model.train(data='/kaggle/input/ioai24-task2-datayolov8/dataset_final/dataset_final', 
                         epochs=10,
                         project="IOAI24-Task2-yolov8",
                         name='test_yolov8_10epochs',
                        batch=32,
                        device=0,
                         imgsz=512,
                        verbose=True,
                        val=True,
                        patience = 10)
```

```python
model2=YOLO("/kaggle/working/IOAI24-Task2-yolov8/test_yolov8_10epochs/weights/best.pt")
data_private = pd.read_csv('/kaggle/input/ioai24-task2-private/private.csv')  
output = pd.read_csv('/kaggle/input/ioai24-task2-private/private.csv')
```

```python
results = model2('/kaggle/input/ioai24-task2-private/private_test/private/images/'+data_private['name'][38])
```

```python
labels = results[0].names
probs = results[0].probs.data.tolist()
best_prediction = np.argmax(probs)
best_prediction_labels = labels[best_prediction]
```

```python
count=0
for i in data_private['name']:
    results = model2('/kaggle/input/ioai24-task2-private/private_test/private/images/'+data_private['name'][count])
    labels = results[0].names
    probs = results[0].probs.data.tolist()
    best_prediction = np.argmax(probs)
    best_prediction_labels = labels[best_prediction]
    output['label'][count] = int(best_prediction_labels)
    count+=1
```

```python
output.to_csv('sample_yolov8.csv', header=True, index=False)
```
